```python
from altaipony.flarelc import FlareLightCurve
from altaipony.lcio import from_mast
from scipy.signal import find_peaks

from matplotlib import pyplot as plt
from astropy import units as u
import pandas as pd
import scipy as sp
import statistics as st

#Energydata = []
Peaktime_and_Energy_sectors51to78 = []

for i in range(54,60):
  flc = from_mast("TIC 233068870" , mode="LC", c=i, mission="TESS", cadence="short", author="SPOC")
  if flc is not None:
    dflc = flc.detrend("savgol")

    norm_de_flux = dflc.detrended_flux / np.nanmedian(dflc.detrended_flux)
    median = np.median(norm_de_flux)
    mad = sp.stats.median_abs_deviation(norm_de_flux)
    times = dflc.time.value
    timeandflux = np.column_stack((times, norm_de_flux))

    #print("normalised detrended flux points : ", norm_de_flux)
    #print("median of normalised detrended flux = ", median)
    #print("median absolute deviation = ", mad)
    #print("time = ", times)

    #timeandflux.shape
    #plt.plot(timeandflux[:,0], timeandflux[:,1])
    #plt.show()

    data = pd.DataFrame(timeandflux,columns=['time','flux'])
    #data = data.set_index('time')

    #data.loc[data.idxmax()]
    #data[data['flux']>median].plot()

    median = np.median(data['flux'])
    x = data['flux'].values
    peaks, _ = find_peaks(x, height= median + 5* mad)

    #peaktime = data['time'][peaks]

    #plt.plot(x)
    #plt.plot(peaks, x[peaks], "x")
    #plt.plot(np.zeros_like(x), "--", color="gray")
    #plt.show()
    #print(peaks)

    data['greater_than_median'] = data['flux']>median

    #data
    #data.head()

 #-------------------------------------------------------------------------------#

    data = data.reset_index()

    ### indexing
    peak_indices = peaks

    list_of_peaks = []

    ### extracting peaks
    def extract_peak(index):
      start_index = index
      end_index = index

      while start_index >= 0 and data['greater_than_median'][start_index]:
         start_index -= 1

      while end_index < len(data) and data['greater_than_median'][end_index]:
         end_index += 1

      if start_index != end_index and (end_index - start_index) >= 3:
         required_df = data[start_index+1:end_index]
         required_df['max_flux'] = required_df['flux'].max()
         required_df['peak_time'] = required_df['time'].loc[required_df['flux'] == required_df['flux'].max()]
         list_of_peaks.append(required_df)


    ### Iterate through the peak indices and extract peaks
    for index in peak_indices:
      extract_peak(index)

 #-------------------------------------------------------------------------------#

    concat_df = pd.concat(list_of_peaks)
    concat_df = concat_df.drop_duplicates(subset=['time'])

    list_of_unique_flares = []
    for max_flux in concat_df['max_flux'].unique():
      list_of_unique_flares.append(concat_df[concat_df['max_flux'] == max_flux])

    #print(len(list_of_unique_flares))
    #for element in list_of_unique_flares:
      #print(element.shape)

    ##uniqueflares = pd.DataFrame(list_of_unique_flares, columns=['time','flux','greater_than_median','maxflux'])

    ### Calculation of Area and Luminosity of all Flare Points in each Flares
    list_of_sum_of_Lflares = []
    peaktime = []
    cadence = []

    for dataframe in list_of_unique_flares:
      dataframe['Aflare'] = pi * (rstar**2) * integral_part * (dataframe['flux'] - 1)
      dataframe['Lflare'] = stefbolt * ((Tflare)**4) * dataframe['Aflare']
      sum_of_Lflares = dataframe['Lflare'].sum()


      ### Pointing out Peak Time for each Flares
      peakt = dataframe['time'].loc[dataframe['flux'] == dataframe['flux'].max()]


      list_of_sum_of_Lflares.append(sum_of_Lflares)
      peaktime.append(peakt.values)


      ### Calculation of Cadence
      flare_duration = (dataframe['time'].max() - dataframe['time'].min()) * 86400
      no_of_flare_points = len(dataframe)
      caden = flare_duration / no_of_flare_points
      cadence.append(caden)

    array_of_sum_of_Lflares = np.array(list_of_sum_of_Lflares)
    array_of_peaktime = np.array(peaktime)
    peaktime_and_sumofLflares = np.column_stack((array_of_peaktime,array_of_sum_of_Lflares))

    ### Calculation of Energy of the Flares
    list_of_Eflares = []

    list_of_Eflares = peaktime_and_sumofLflares[:,1] * cadence
    #final = np.column_stack((peaktime_and_sumofLflares[:,0],list_of_Eflares))
    sector = int(i)*(np.ones(len(list_of_Eflares),dtype = int))
    peakandenergy = np.column_stack((peaktime_and_sumofLflares[:,0],list_of_Eflares,sector))
    Peaktime_and_Energy_sectors51to78.append(peakandenergy)



  #energy = np.concatenate(Energy, axis=0)
  #peaktime_and_energy = np.column_stack((peaktime,energy))
```
